```python
import cv2
import matplotlib.pyplot as plt
```

```python
img=cv2.imread('/content/dog.webp')
img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.figure(figsize=(8,4))
plt.imshow(img)
plt.axis('off')
plt.title('Original image')
```

```python
gray_img=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
plt.figure(figsize=(8,4))
plt.imshow(gray_img,cmap='gray')
plt.axis('off')
plt.title('Gray image')
```

```python
inverted=cv2.bitwise_not(gray_img)
plt.figure(figsize=(8,4))
plt.imshow(inverted,cmap='gray')
plt.axis('off')
plt.title('Inverted image')
```

```python
Gsn_image = cv2.GaussianBlur(inverted,(7,9),sigmaX=0,sigmaY=0)
plt.figure(figsize=(8,6))
plt.imshow(Gsn_image,cmap='gray')
plt.axis('off')
plt.title('Gaussian Blur Image')
plt.show()
```

```python
pencil_sktch=cv2.divide(gray_img,255-Gsn_image,scale=255)
plt.figure(figsize=(10,7))
plt.imshow(pencil_sktch,cmap='gray')
plt.axis('off')
plt.title('Pencil Sketch')
plt.show()
```
